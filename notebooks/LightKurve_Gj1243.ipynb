Following LightKurve tutorial example to make sure things are working :

First, we start Python and use the search_targetpixelfile function to obtain the Kepler pixel data for the star from the data archive:

( http://docs.lightkurve.org/tutorials/)

```python
from lightkurve import search_targetpixelfile
pixelfile = search_targetpixelfile(8462852, quarter=16).download(quality_bitmask='hardest');
```

```python
#Display the first image in this data set
pixelfile.plot(frame=1);
```

```python
#we can extract a lightcurve by simply summing up all the pixel values in each image:
lc = pixelfile.to_lightcurve(aperture_mask='all');

#The above method returned a KeplerLightCurve object which gives us access to the flux over time, 
#which are both available as array objects. 
#The time is in units of days and the flux is in units electrons/second.
lc.time, lc.flux
```

```python
#plot the light curve
lc.plot();
```

Following the tutorial for how to make a full light curve with all of the quarters of data

```python
#one big light curve

from lightkurve import search_lightcurvefile


target = 'Kepler-8b'
```

```python
print(target)
```

```python
lc = search_lightcurvefile(target, quarter=1).download().PDCSAP_FLUX
for q in range(2,17):
    lc = lc.append(search_lightcurvefile(target, quarter=q).download().PDCSAP_FLUX)
```

```python
lc.scatter();
```

```python
#normalizing for systematics

lc = search_lightcurvefile(target, quarter=1).download().PDCSAP_FLUX.normalize()
for q in range(2,8):
    lc = lc.append(search_lightcurvefile(target, quarter=q).download().PDCSAP_FLUX.normalize())
```

```python
lc.scatter();
```

```python
# removing long term trends
lc = search_lightcurvefile(target, quarter=1).download().PDCSAP_FLUX.normalize().flatten(window_length=401)
for q in range(2,8):
    lc = lc.append(search_lightcurvefile(target, quarter=q).download().PDCSAP_FLUX.normalize().flatten(window_length=401))
```

```python
lc.scatter();
```

```python
lc.fold(period=3.52254, t0=1.35).bin().scatter();
```
